# Importáljuk a megfelelő packageket

```python
import matplotlib.pyplot as plt
import numpy
from sklearn.metrics import r2_score
```

## 3. Polinomális regresszió

**Ismertető**  
A lineáris regresszió kiterjesztése: ugyanúgy lineárisan becsüli a paramétereket, de a bemeneti változó(k) magasabb hatványait (\(x^2, x^3, \dots\)) is felveszi. Így **görbe**, de mégis sima (polinom) vonalat tud illeszteni egy‑egy változó mentén.

**Szóbeli példa**  
Egy közlekedésbiztonsági szakértő vizsgálja, hogyan nő a **féktávolság** (\(y\)) az **autó sebességének** (\(x\)) függvényében. Tapasztalatból tudják, hogy a távolság nem lineárisan, hanem nagyjából a sebesség négyzetével arányosan nő, ezért egy másodfokú (kvadratikus) polinomot illesztenek.


```python
dataset = numpy.loadtxt("adatok/CO2.csv", delimiter=",")
# bemeneti értékek az első adat év (X), a mért co2 szint(Y)
x = dataset[:,0]
y = dataset[:,1]
#NumPy polynomiális model készítése:
elorejelzo_model = numpy.poly1d(numpy.polyfit(x, y, 3))
#illeszkedés mértéke
print(r2_score(y, elorejelzo_model (x)))
#légköri CO2 érték vizsgálata adott értékhatárokhoz. A kérdés, melyik évben érjük el az adott szintet.
for CO2 in range(2025,6000):
    if elorejelzo_model(CO2)<1400:
        CO2_1400=CO2
    if elorejelzo_model(CO2) < 2000:
        CO2_2000 = CO2
    if elorejelzo_model(CO2)<5000:
        CO2_5000=CO2
    if elorejelzo_model(CO2)<10000:
        CO2_10000=CO2
    if elorejelzo_model(CO2)<40000:
        CO2_40000=CO2
    if elorejelzo_model(CO2)<100000:
        CO2_100000=CO2
    if elorejelzo_model(CO2)<200000:
        CO2_200000=CO2
    if elorejelzo_model(CO2)<1000000:
        CO2_1000000=CO2

print("1400 ppm, Figyelemzavar",CO2_1400)
print("2000 ppm, Stressz, kalcium csontbaépülésének zavara, vérkeringési és veseműködésizavar ",CO2_2000)
print("5000 ppm, Foglalkozás-egészségügyi határérték, fejfájás ",CO2_5000)
print("10000 ppm, Zárt helységben megengedett legnagyobb érték",CO2_10000)
print("40000 ppm, Kilélegzett levegő",CO2_40000)
print("100000 ppm, Gyertya elalszik, hányinger, eszméletvesztés ",CO2_100000)
print("200000 ppm, fulladás általi halál",CO2_200000)
print("1000000 ppm, Teljes telítettség",CO2_1000000)
```

# Készítsünk grafikont a a CO2 elorejelzésre

```python
#Grafikon készítés
illesztett_gorbe = numpy.linspace(1950, 2030, 200)
plt.title('CO2 szint évente')
plt.xlabel('Év')
plt.ylabel('CO2')
plt.scatter(x, y)
plt.plot(illesztett_gorbe, elorejelzo_model (illesztett_gorbe))
plt.show()
```
