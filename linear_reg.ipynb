# Import packages

```python
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

## 1. Egyszerű lineáris regresszió

**Ismertető**  
Egyetlen magyarázó változóval (\(x\)) becsli a célváltozót (\(y\)). A kapcsolatot egy egyenes (lineáris trend) írja le. A modell akkor jó választás, ha az adatok pontfelhője ránézésre is közel egyenest követ.

**Szóbeli példa**  
Képzeld el, hogy egy fagylaltárus szeretné megtudni, hogyan befolyásolja a **napi átlaghőmérséklet** (\(x\)) a **naponta eladott gombócok számát** (\(y\)). A feltételezés az, hogy melegebb napokon több fagylalt fogy lineárisan növekvő ütemben.

```python
# Dataset
data = {
    "negyzetmeter": [200, 250, 300, 450, 500],
    "ar": [1550000, 2100000, 2700000, 3150000, 3500000]
}

df = pd.DataFrame(data)

# Független változó (independent variable)
x = df[["negyzetmeter"]]

# Függő változó (dependent variable)
y = df[["ar"]]
model = LinearRegression()

# Tanítjuk a modellt
model.fit(x,y)
```

# Készítünk egy előrejelzést próbának

```python
#előrejelzés
negyzetmeter_input=[[600]] # azért van [[]] mert a .predict function csak array-like dolgot fogad el.
elorejelzett_ar=model.predict(negyzetmeter_input)
print("Előrejelzett ár", negyzetmeter_input,"négyzetméteres házra:", elorejelzett_ar[0])
```

## Megjelenítjük a Lineáris Regresszíót grafikonon

```python
#Grafikon
plt.scatter(x,y,color="blue", label="Aktuális árak")
plt.plot(x, model.predict(x),color="red", label="Regressziós egyenes")
plt.xlabel("Négyzetméter")
plt.ylabel("Ár")
plt.title("Ház ára a négyzetméter függvényében")
plt.legend()
plt.grid(True)
plt.show()
```
