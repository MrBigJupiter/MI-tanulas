```python
import numpy as np

x=[]
y=[]

# INPUT
I1=2
I2=1
I3=2


#Súlyok
w1=0.2
w2=0.3
w3=0.1
w4=0.1
w5=0.4
w6=0.28
w7=0.21
w8=0.34
w9 = 0.15
w10 = 0.42
weights = np.array([
    0.2,   # w1
    0.3,   # w2
    0.1,   # w3
    0.1,   # w4
    0.4,  # w5
    0.28,  # w6
    0.21,   # w7
    0.34,   # w8
    0.15,  # w9
    0.42    # w10
])

#tanitasi rata
a=0.001
#output elvárt érték O1, O2
O1=2
O2=1
for i in range(100):
    H2=I1*w1 + I3*w4
    H1= I2*w2
    H3=I2*w3 + I3*w5
    O_calc_1= H1*w6 + H3*w9
    O_calc_2= H1*w7 + H2*w8 + H3*w10
    lambda_test = 0.005
    print("Iteráció:",i+1)
    print("Elvárt output1:",O1,"eredmény output1:",O_calc_1)
    print("Elvárt output2:", O2, "eredmény output2:", O_calc_2)
    hiba = (0.5*(O1 - O_calc_1)**2 + 0.5*(O2 - O_calc_2)**2)/2
    regularization = (lambda_test / 2) * np.sum(weights ** 2)
    hiba = (hiba + regularization)
    print("A hiba mértéke:",hiba)
    #sulyok értékének a módosítása
    w1 = w1-a*(hiba/w1)
    w2 = w2-a*(hiba/w2)
    w3 = w3-a*(hiba/w3)
    w4 = w4-a*(hiba/w4)
    w5 = w5-a*(hiba/w5)
    w6 = w6-a*(hiba/w6)
    w7 = w7-a*(hiba/w7)
    w8 = w8-a*(hiba/w8)
    w9 = w9-a*(hiba/w9)
    w10 = w10-a*(hiba/w10)
    x.append(i+1)
    y.append(hiba)
    print(x)
    print(y)
print(f"Hiba mértéke: {hiba}")
print(f"100 iteráció után ez az 1. output értéke: {O_calc_1}")
print(f"100 iteráció után ez a 2. output értéke: {O_calc_2}")
```
